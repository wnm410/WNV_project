```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import roc_auc_score, auc, roc_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import PolynomialFeatures

pd.set_option('display.max_columns', 500)
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

Reading in train and test datasets, resetting the date as index

```python
df = pd.read_csv('train_weather_per_station.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace = True)
df.head()
```

```python
df.shape
```

```python
test = pd.read_csv('test_weather_per_station.csv')
test['Date'] = pd.to_datetime(test['Date'])
test.set_index('Date', inplace = True)
test.head()
```

```python
print(f'Number of columns in the train set: {df.shape[1]}')
print(f'Number of columns in the test set: {test.shape[1]}')
```

Since there is a column of Unspecified Species in the test data that is not in the train data, I substitute that species in the test data with the most popular species.

```python
# checking which is the most popular species 

test[['CULEX ERRATICUS','CULEX PIPIENS','CULEX PIPIENS/RESTUANS','CULEX RESTUANS','CULEX SALINARIUS',
     'CULEX TARSALIS','CULEX TERRITANS']].sum()
```

```python
# assigning culex pipiens/restuans to the unspecified culex
for index, row in test.iterrows():
    if row['UNSPECIFIED CULEX'] == 1:
        test.loc[index,'CULEX PIPIENS/RESTUANS'] = 1

# check if the number of culex pipiens/restuans has actually increased
test['CULEX PIPIENS/RESTUANS'].sum()
```

```python
df.head()
```

```python
# droppig the unspecified culex
test.drop('UNSPECIFIED CULEX', axis = 1, inplace = True)
test.head()
```

Tried to build implemet lagged features but wasn't able to finish so I ended up not implementing them.

```python
def buildLaggedFeatures(df, lag):  # df - dataframe, lag - list of numbers defining lagged values. Builds lagged weather features 
    new_dict = {}
    for col_name in df:
        new_dict[col_name] = df[col_name]
        # create lagged Series
        for l in lag:
            new_dict['%s_lag%d' %(col_name,l)] = df[col_name].shift(l)
    res = pd.DataFrame(new_dict,index=df.index)
    return res
```

```python
weather_feats = ['Tmax','Tmin','Tavg','Depart','DewPoint','WetBulb','Cool','PrecipTotal','ResultSpeed','ResultDir']
lagged_train = buildLaggedFeatures(df[weather_feats], [1,3,5,7])
lagged_train.drop(columns = weather_feats, inplace = True)

lagged_test = buildLaggedFeatures(test[weather_feats], [1,3,5,7])
lagged_test.drop(columns = weather_feats, inplace = True)
lagged_train.head()
```

# Defining X and y, and train/test split

Oversampling the positive class through bootstrapping

```python
df.shape
```

```python
df['WnvPresent'].value_counts()
```

```python
# features = [feat for feat in df.columns if feat != "WnvPresent"]

X = df.drop(columns = ['WnvPresent'])
y = df['WnvPresent']
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 17, stratify = y)
```

Feature engineering through `PolynomialFeatures`

```python
X.head()
```

```python
weather_feats = ['Tmax','Tmin','Tavg','Depart','DewPoint','WetBulb','Cool','PrecipTotal','ResultSpeed','ResultDir']

poly = PolynomialFeatures(2)
poly_feat = pd.DataFrame(poly.fit_transform(X[weather_feats]), columns = poly.get_feature_names(X[weather_feats].columns))
poly_feat.drop(labels=weather_feats, axis=1, inplace=True)
```

```python
poly_feat.shape
```

```python
poly_feat.set_index(df.index, inplace = True)
poly_feat.drop(columns = ['1'],inplace = True)
poly_feat.head()
```

```python
df = df.join(poly_feat)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 17)
```

# KNN model

```python
k = np.arange(3,30,2)
parameters = {'n_neighbors': k,
              'p': [1,2] }
knn = KNeighborsClassifier()
clf = GridSearchCV(knn,parameters,cv=10)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
auc_knn = roc_auc_score(y_test, y_pred)
auc_knn
```

# Logistic Regression model

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
auc_logreg = roc_auc_score(y_test, y_pred)
auc_logreg
```

# Random Forest w/ Grid Search

```python
params = {'n_estimators': [170,180,190],
          'max_depth': [145,150,155]}
rfc = RandomForestClassifier(class_weight='balanced') 
rfcl = GridSearchCV(rfc,params,cv=10)
rfcl.fit(X_train, y_train)
rfcl.best_params_
```

```python
y_pred = rfcl.predict(X_test)
auc_rfcl = roc_auc_score(y_test, y_pred)
auc_rfcl
```

```python
probs = rfcl.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, preds)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8,8))
plt.title('Receiver Operating Characteristic', ha = 'left', x = 0, fontsize = 20)
plt.plot(fpr, tpr, c = 'mediumslateblue', label = 'AUC = %0.2f' % roc_auc, lw = 4)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--', lw = 4)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate', fontsize = 18)
plt.xlabel('False Positive Rate', fontsize = 18)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18);
```
