```python
import pandas as pd
import numpy as np

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
data = pd.read_csv('./data/train.csv')
```

```python
data['WnvPresent'].value_counts()
```

```python
rdata = data.groupby('Trap')['NumMosquitos'].agg(['mean','count'])
print(rdata)
```

```python
print(rdata.sort_values(by='count',ascending=False))[:10]
```

```python
print(rdata.sort_values(by='mean',ascending=False))[:10]
```

```python
blockdata = data.groupby('Block')['NumMosquitos'].agg(['mean','count'])
print(blockdata.sort_values(by='count',ascending=False))[:10]
```

```python
print(blockdata.sort_values(by='mean',ascending=False))[:10]
```

```python
wnvdata = data.groupby('Trap')['WnvPresent'].agg(['mean','count'])
print(wnvdata.sort_values(by='count',ascending=False))[:10]
```
