```python
from math import sin, cos, sqrt, asin, radians
def distance(lat1,lon1,lat2,lon2):
    # transforming longitude and latitude to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    # latitude and longitude difference
    dlat = lat2-lat1
    dlon = lon2-lon1
    
    # radius of the Earth in km
    R = 6373 
    
    # complicated multiplier calculated separately
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    
    # Haversine formula
    return 2*asin(sqrt(a))*R
```

```python
#dates that are right before and right after sprays (if 3 entries, the middle netry is the date of spray)
spray_dict = {}

spray_dict['2011-08-29'] = ['2011-08-26', '2011-09-01']
spray_dict['2011-09-07'] = ['2011-09-02', '2011-09-12']
spray_dict['2013-07-17'] = ['2013-07-12', '2013-07-19']
spray_dict['2013-07-25'] = ['2013-07-19', '2013-07-25', '2013-08-01']
spray_dict['2013-08-08'] = ['2013-08-01', '2013-08-08', '2013-08-15']
spray_dict['2013-08-15'] = ['2013-08-08', '2013-08-15', '2013-08-22']
spray_dict['2013-08-16'] = ['2013-08-15', '2013-08-22']
spray_dict['2013-08-22'] = ['2013-08-15', '2013-08-22', '2013-08-29']
spray_dict['2013-08-29'] = ['2013-08-22', '2013-08-29', '2013-09-06']
spray_dict['2013-09-05'] = ['2013-08-29', '2013-09-06']
```

```python
spray_dict
```

```python
for k, v in spray_dict.items():
    spray_describe = spray[spray['Date'] == k].describe()
    mask1_1 = data['Longitude'] >= spray_describe.loc['min', 'Longitude'] 
    mask1_2 = data['Longitude'] <= spray_describe.loc['max', 'Longitude']
    mask2_1 = data['Latitude'] >= spray_describe.loc['min', 'Longitude']
    mask2_2 = data['Longitude'] <= spray_describe.loc['max', 'Longitude']
    big_mask = mask1_1 & mask1_2 & mask2_1 & mask2_2
    if len(v) == 2:
        spraytestbefore = data[data['Date'] == v[0]]
        spraytestafter = data[data['Date'] == v[1]]
        if len(spraytestbefore[big_mask]) == 0:
            print('Spray date: ', k, 'No data before')
        else:
            print('Spray date: ', k,
                  'Mozzies before: ', spraytestbefore[big_mask]['NumMosquitos'].sum(),
                  'Mozzies after: ', spraytestafter[big_mask]['NumMosquitos'].sum())
    elif len(v) == 3:
        spraytestbefore = data[data['Date'] == v[0]]
        spraytestduring = data[data['Date'] == v[1]]
        spraytestafter = data[data['Date'] == v[2]]
        if len(spraytestbefore[big_mask]) == 0:
            print('Spray date: ', k, 'No data before')
        else:
            print('Spray date: ', k,
                  'Mozzies before: ', spraytestbefore[big_mask]['NumMosquitos'].sum(),
                  'Mozzies during: ', spraytestduring[big_mask]['NumMosquitos'].sum(),
                  'Mozzies after: ', spraytestafter[big_mask]['NumMosquitos'].sum())
```

```python
for k, v in spray_decay.items():
    spray_describe = spray[spray['Date'] == k].describe()
    mask1_1 = data['Longitude'] >= spray_describe.loc['min', 'Longitude'] 
    mask1_2 = data['Longitude'] <= spray_describe.loc['max', 'Longitude']
    mask2_1 = data['Latitude'] >= spray_describe.loc['min', 'Longitude']
    mask2_2 = data['Longitude'] <= spray_describe.loc['max', 'Longitude']
    big_mask = mask1_1 & mask1_2 & mask2_1 & mask2_2
    
    to_graph = pd.DataFrame(columns = ['Date', 'Num_Mosq', 'Precip'])
    
    for i in range(len(v)):
        spraytest = data[data['Date'] == v[i]]
        weather_mask1 = weather['Station'] == 2
        weather_mask2 = weather['Date'] == v[i]
        to_graph = to_graph.append({'Date': v[i], 
                                    'Num_Mosq': spraytest[big_mask]['NumMosquitos'].sum(), 
                                    'Precip': float(weather[(weather_mask1) & (weather_mask2)]['PrecipTotal'].values[0])}, 
                                   ignore_index=True)
        
    to_graph['Date'] = pd.to_datetime(to_graph['Date'])
    
    fig, ax1 = plt.subplots(figsize = (6,3))
    
    plt.title('{} Spray'. format(k), fontsize = 16)
    plt.xticks(rotation = 45)
    
    ax1.plot(to_graph['Date'], to_graph['Num_Mosq'], color = 'slategray', lw = 2)
    ax1.axvline(x=k, color='k', linestyle='--')
    ax1.set_ylabel('Number of Mosquitos', color = 'slategray', fontweight = 'bold')
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('Total Precipitation (in)', color = 'royalblue', fontweight = 'bold')
    ax2.plot(to_graph['Date'], to_graph['Precip'], color = 'royalblue', lw = 2)
    ax2.text(to_graph['Date'].iloc[-1], to_graph['Precip'].max()-to_graph['Precip'].mean(), '--- Spray Date', horizontalalignment='right')
```
